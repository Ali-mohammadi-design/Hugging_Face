```python
from PIL import Image
```

Image is a tool that allow you to open up the images

We can convert this Image to a array (numbers)

```python
import numpy as np
```

```python
pic_array=np.asarray(pic)
```

```python
#pic_array
```

Note: As is clear here all the values are between 0 to 255. If you devide the array to 255 it will be normalized between 0 and 1.

```python
pic_array.shape
```

Note: 1024 is width of the array, the second 1024 is height of the array. 4 is the dimension of the arrage. we have one layer for Red, one layer for grenn and one layer for blue, the fourth layer for the transparency. Which means if this picture lay on anouther picture how much it is tranparent.

Note: We can show the arrays with numpy

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(pic_array)
```

Note: As is clear the width and height are 1024

As was told we have three color chanels here:

```python
pic_red=pic_array.copy()
```

```python
pic_red[:,:,1]=0
#Hight and wedth is every thing but we just choose the layer 1 and put it on zero
```

```python
plt.imshow(pic_red)
```

As is clear we do not have the green values any more. This is just red and blue!

```python
pic_red[:,:,2]=0
plt.imshow(pic_red)
```

```python
import cv2
```

```python
img=cv2.imread('/content/puppy.png')
```

Generate an array

```python
#img
```

```python
plt.imshow(img)
```

Note: The picture is different because the arrangement of RGB in this package is different. It is Blue Green Red

```python
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

Note: you can also show the image on the gray scale

```python
img_gray=cv2.imread('/content/puppy.png',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray')
```

Note: you can also resize the images!

```python
img_resize=cv2.resize(img_rgb,(512,1024))
plt.imshow(img_resize)
```
