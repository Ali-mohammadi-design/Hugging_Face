```python
pipe_1=pipeline(task='text-classification')
```

```python
pipe_1('this boy is nice!')
```

If, for instance, we want to classify some financial text we can explore the models in hugging face and find a specific model, if it is available.

```python
pipe_2=pipeline(task='text-classification', model='prosusAI/finbert')
```

```python
qa_bot=pipeline('question-answering')
```

```python
text='''Iran,[a] officially the Islamic Republic of Iran (IRI),[b] also known as Persia,[c] is a country in West Asia.
 It borders Turkey to the northwest and Iraq to the west, Azerbaijan, Armenia, the Caspian Sea, and Turkmenistan to the north, Afghanistan to the east, Pakistan to the southeast, the Gulf of Oman and the Persian Gulf to the south.
 With a mostly Persian-ethnic population of almost 90 million in an area of 1,648,195 km2 (636,372 sq mi), Iran ranks 17th globally in both geographic size and population.
  It is the sixth-largest country entirely in Asia and one of the world's most mountainous countries. Officially an Islamic republic, Iran has a Muslim-majority population.
   The country is divided into five regions with 31 provinces. Tehran is the nation's capital, largest city and financial center.'''
```

```python
question='what is Iran capital?'
```

```python
qa_bot(context=text, question=question)
```
