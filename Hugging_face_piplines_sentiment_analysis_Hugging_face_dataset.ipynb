What is the difference among data set and dataframe?


A dataset is a collection of data. It is a broad term that can refer to any structured collection of data, including data stored in a database, a CSV file, or any other format.


A dataframe is a specific type of data structure used primarily for data analysis and manipulation. It is typically used to represent tabular data (i.e., data in rows and columns).

**Using Piplines**

```python
from transformers import pipeline
```

In this level you have to go to the hugging face and find a model which is suitable to you.
 on the model page,  top right of the page, use this model section,  you can just copy the pipeline code to bring the model here by the pipline.

Evenif you do not know any model you can just mention the task that you want, like the following code, and the Hugging face selects a default model for you.

```python
classifier=pipeline('sentiment-analysis')
```

```python
classifier('this is a great president!')
```

Let's try the classifier!!

```python
test_df=reviews['train'].to_pandas()
```

```python
test_df.head()
```

```python
test_df['text'].head()[0]
```

Note: if we apply the classifier here it returns a list of dictionary including score and label. However, we only need the label now. Thus we can use a function!

```python
def score(text):
  return classifier(text)[0]['label']
```

```python
test_df['model_prediction']=test_df['text'].apply(score)
```
